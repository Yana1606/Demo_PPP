## Обработка вещественных признаков (заполнение пропусков) <a id='toc1_4_'></a>

Проверим количество пропущенных значений в каждом столбце для их дальнейшего заполнения.

```python
missing_values = X_train.isnull().sum()
```

```python
columns_with_missing = missing_values[missing_values > 0].index.tolist()
print(f"Столбцы с пропусками: {columns_with_missing}")
```

Воспользуемся статистическими методами для заполнения пропущенных значений и найдем лучший метод для заполнения по KL-дивергенции.

```python
def calculate_kl_divergence(original_series, filled_series, bins=20):

    original_hist, bin_edges = np.histogram(original_series.dropna(), bins=bins, density=True)
    filled_hist, _ = np.histogram(filled_series, bins=bin_edges, density=True)
    
    original_hist = original_hist + 1e-10
    filled_hist = filled_hist + 1e-10
    
    kl_divergence = entropy(original_hist, filled_hist)
    return kl_divergence
```

```python
X_train_statistical = X_train.copy()
```

```python
def interpret_kl_divergence(kl_value):
    
    if kl_value < 0.01:
        return "Очень похожие распределения"
    elif kl_value < 0.05:
        return "Похожие распределения"
    elif kl_value < 0.1:
        return "Умеренное различие"
    elif kl_value < 0.5:
        return "Заметное различие"
    else:
        return "Сильное различие"
```

```python
for col in columns_with_missing:
    print(f"\nАнализ столбца: {col}")
    print(f"Количество пропусков: {X_train[col].isnull().sum()}")
    print(f"Процент пропусков: {X_train[col].isnull().mean():.2%}")
    
    # Возможные значения для заполнения
    fill_values = {
        'mean': X_train[col].mean(),
        'median': X_train[col].median(),
        'mode': X_train[col].mode()[0] if not X_train[col].mode().empty else X_train[col].median()
    }
    
    # Создаем отдельную фигуру для каждого столбца
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle(f'Анализ заполнения пропусков: {col}', fontsize=16, fontweight='bold')
    
    # 1. Исходное распределение (без пропусков)
    ax1 = axes[0, 0]
    original_data = X_train[col].dropna()
    ax1.hist(original_data, bins=20, alpha=0.7, color='blue', edgecolor='black')
    ax1.set_title('Исходное распределение (без пропусков)')
    ax1.set_xlabel('Значение')
    ax1.set_ylabel('Частота')
    ax1.grid(True, alpha=0.3)

    stats_text = f"Среднее: {original_data.mean():.2f}\nМедиана: {original_data.median():.2f}\nМода: {original_data.mode()[0]:.2f}"
    ax1.text(0.02, 0.98, stats_text, transform=ax1.transAxes, verticalalignment='top',
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
    
    kl_results = {}

    methods = ['mean', 'median', 'mode']
    colors = ['red', 'green', 'orange']
    
    for i, (method, fill_value) in enumerate(fill_values.items()):

        filled_series = X_train[col].fillna(fill_value)
        
        kl_divergence = calculate_kl_divergence(X_train[col].dropna(), filled_series)
        kl_results[method] = kl_divergence
        
        ax = axes[(i+1)//2, (i+1)%2]  
        ax.hist(filled_series, bins=20, alpha=0.7, color=colors[i], edgecolor='black', 
               label=f'Заполнено ({method})')
        ax.hist(original_data, bins=20, alpha=0.5, color='blue', edgecolor='black', 
               label='Исходное')
        ax.set_title(f'Заполнено {method}\nKL-дивергенция: {kl_divergence:.4f}')
        ax.set_xlabel('Значение')
        ax.set_ylabel('Частота')
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

    print(f"\n  Анализ KL-дивергенции для {col}:")
    for method, kl in kl_results.items():
        interpretation = interpret_kl_divergence(kl)
        print(f"    {method}: {kl:.4f} - {interpretation}")

    best_method = min(kl_results, key=kl_results.get)
    best_value = fill_values[best_method]
    best_kl = kl_results[best_method]
    
    print(f"     Лучший метод: {best_method}")
    print(f"     Значение для заполнения: {best_value:.4f}")
    print(f"     KL-дивергенция: {best_kl:.4f}")
    print(f"     Интерпретация: {interpret_kl_divergence(best_kl)}")

    X_train_statistical[col] = X_train[col].fillna(best_value)
    
    print(f"Пропуски в столбце '{col}' заполнены методом '{best_method}'")
```

### Оценка качества заполнения пропусков машинным методом

**NRMSE** — это нормализованная среднеквадратичная ошибка. Она показывает, насколько сильно восстановленные (заполненные) значения отличаются от истинных.

```python
def calculate_nrmse(original, imputed, mask):
    
    masked_original = original[mask]
    masked_imputed = imputed[mask]
    
    if len(masked_original) == 0:
        return np.nan
    
    mse = np.mean((masked_original - masked_imputed) ** 2)
    rmse = np.sqrt(mse)
    nrmse = rmse / (np.max(original) - np.min(original))
    
    return nrmse
```

```python
clean_data = X_train.dropna().copy()
print(f"Размер чистого датасета: {clean_data.shape}")
```

```python
np.random.seed(42)
masked_data = clean_data.copy()
artificial_mask = {}
```

Убираем все пропуски из обучающей выборки и заполняем ее "искусственными" пропусками.

```python
for col in clean_data.columns:
    if clean_data[col].dtype in ['float64', 'int64']:

        n_missing = int(0.2 * len(clean_data))
        missing_indices = np.random.choice(clean_data.index, n_missing, replace=False)
        
        artificial_mask[col] = clean_data.loc[missing_indices, col].copy()

        masked_data.loc[missing_indices, col] = np.nan

print("Созданы искусственные пропуски (20% данных)")
```

Применяем IterativeImputer для искусственных пропусков.

```python
evaluation_imputer = IterativeImputer(max_iter=10, random_state=42)
imputed_data = evaluation_imputer.fit_transform(masked_data)
imputed_data = pd.DataFrame(imputed_data, columns=masked_data.columns, index=masked_data.index)
```

Считаем средний NRMSE по каждому признаку и выводим получившийся результат.

```python
nrmse_results = {}
for col in clean_data.columns:
    if col in artificial_mask:
        col_mask = masked_data[col].isna()
        if col_mask.sum() > 0:
            nrmse = calculate_nrmse(clean_data[col], imputed_data[col], col_mask)
            nrmse_results[col] = nrmse

print("\nРезультаты оценки NRMSE:")
for col, nrmse in nrmse_results.items():
    print(f"  {col}: NRMSE = {nrmse:.6f}")
```

```python
if nrmse_results:
    plt.figure(figsize=(10, 6))
    plt.bar(nrmse_results.keys(), nrmse_results.values(), color='skyblue')
    plt.title('Качество заполнения пропусков (NRMSE)')
    plt.xlabel('Признаки')
    plt.ylabel('NRMSE')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

Почти все признаки имеют NRMSE < 0.01, что означает, что IterativeImputer восстановил значения очень точно.

## Детекция выбросов и аномалий в данных <a id='toc1_6_'></a>

### Поиск выбросов в данных с помощью **IQR** и **Robust Z-score.**

Коэффициент **k=1.5** для IQR - это стандартный коэффициент в статистике для умеренных выбросов.

В Robust Z-score **threshold=3** соответствует правилу трех сигм в статистике. Robust Z-score использует MAD вместо std, что делает его устойчивее к выбросам.

```python
X = X_train_machine.copy()

def find_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers
```

```python
def find_outliers_robust_zscore(data, threshold=3):
    median = data.median()
    mad = np.median(np.abs(data - median))  
    if mad == 0:
        mad = np.std(data)
    robust_zscore = np.abs(0.6745 * (data - median) / mad)
    outliers = robust_zscore > threshold
    return outliers
```

### Isolation Forest

Isolation Forest — это алгоритм, который ищет аномалии по принципу "чем страннее точка, тем проще её отличить от остальных".

Шаги алгоритма:

- Берём случайные кусочки данных и строит много простых "деревьев", где в каждом узле данные случайно разделяются по какому-то признаку.

- Аномальные точки изолируются быстро — обычно за 1-2 разделения, потому что они сильно отличаются от основной массы и их легко "отсеять".

- Нормальные точки требуют много разделений — они похожи друг на друга, поэтому их приходится долго "разбирать" по кучкам.

Алгоритм просто считает, сколько в среднем шагов нужно чтобы изолировать каждую точку. Те, что изолировались быстро — аномалии. Те, что долго — нормальные данные.

contamination - доля выбросов в данных

```python
iso_forest = IsolationForest(contamination=0.1, random_state=42)
iso_forest_pred = iso_forest.fit_predict(X_robust)
iso_forest_anomalies = set(X_train_cleaned_outliers[iso_forest_pred == -1].index)
```

### One-class SVM

One-Class SVM — это алгоритм обнаружения аномалий, который работает по принципу "обучиться на нормальном и отсечь всё ненормальное".

В отличие от обычных классификаторов, One-Class SVM обучается только на нормальных данных и строит границу, отделяющую эту нормальную область от всего остального пространства.

Шаги алгоритма:

- Преобразуем данные в пространство высшей размерности с помощью ядерной функции (RBF), где нормальные данные становятся более "сгруппированными"

- Находим гиперплоскость (границу), которая отделяет большинство нормальных точек от начала координат в этом новом пространстве

- Стараемся сделать эту границу как можно более плотной вокруг нормальных данных

Ключевой параметр — nu (ν):

- Определяет, какой процент точек считается аномалиями (например, ν=0.1 означает, что 10% данных будут считаться выбросами)

- Контролирует компромисс между плотностью границы и допущением ошибок

```python
one_class_svm = OneClassSVM(nu=0.1, kernel='rbf', gamma='scale')
one_class_svm_pred = one_class_svm.fit_predict(X_robust)
one_class_svm_anomalies = set(X_train_cleaned_outliers[one_class_svm_pred == -1].index)
```

### DBSCAN

DBSCAN — это алгоритм кластеризации, который заодно отлично находит аномалии. Его главная идея: "нормальные данные образуют плотные скопления, а выбросы — это всё, что стоит особняком".

Алгоритм ищет области пространства, где точки расположены плотно, и объявляет их кластерами. Всё, что не попадает в эти плотные области, считается шумом (аномалиями).

Ключевые параметры:

- eps (ε) — радиус окрестности вокруг точки

- min_samples — минимальное количество точек в ε-окрестности

Шаги алгоритма:

- Для каждой точки проверяет, сколько соседей попадает в её ε-окрестность

- Если соседей достаточно (≥ min_samples) — это "ядровая" точка, и она становится частью кластера

- Все точки в окрестностях ядерных точек тоже добавляются в кластер

- Точки, у которых мало соседей И которые не попали в окрестности ядерных точек, объявляются шумом (аномалиями)

```python
dbscan = DBSCAN(eps=0.2, min_samples=25)
dbscan_pred = dbscan.fit_predict(X_robust)
dbscan_anomalies = set(X_train_cleaned_outliers.loc[dbscan_pred == -1].index)
```

# Выбор моделей ML и метрик <a id='toc3_'></a>

1. **CatBoost (Categorical Boosting)**

CatBoost — это продвинутая реализация градиентного бустинга, специально разработанная для эффективной работы с категориальными признаками. В отличие от других алгоритмов, требующих предварительного кодирования категориальных переменных, CatBoost обрабатывает их автоматически с помощью Ordered Boosting — техники, которая предотвращает переобучение за счет использования перестановок для вычисления статистик.

В задачах с потенциально смешанными типами данных (числовые и категориальные признаки) CatBoost часто показывает лучшие результаты благодаря своей способности корректно обрабатывать категории без потери информации. Его алгоритм упорядоченного бустинга также обеспечивает хорошую устойчивость к переобучению.

$
x_{i,k} = (∑_{j=1}^{p-1} [x_{σ_j, k} = x_{σ_i, k}] * Y_{σ_j} + a * P) / (∑_{j=1}^{p-1} [x_{σ_j, k} = x_{σ_i, k}] + a)
$

Где σ — случайная перестановка объектов, p — текущий объект, a — коэффициент сглаживания, P — априорное значение.

2. **LightGBM (Light Gradient Boosting Machine)**

LightGBM использует инновационный подход leaf-wise роста деревьев (в отличие от level-wise в традиционном GBDT) и histogram-based алгоритмы для группировки значений признаков в бины. Это делает его исключительно быстрым и эффективным по памяти, особенно на больших наборах данных.

Для исследовательских задач, где нужно быстро перебирать множество конфигураций, скорость LightGBM является критическим преимуществом. Его эффективность на данных с большим количеством признаков также хорошо подходит для современных датасетов.

$
A = {x_i ∈ X : |∇L| топ-а}, B = {x_i ∈ X : случайные b из оставшихся}
$

Где ∇L — градиент функции потерь.

3. **GradientBoosting (Scikit-learn реализация)**

Классический градиентный бустинг, который последовательно строит деревья, где каждое следующее дерево обучается на ошибках предыдущих. Это надежная, проверенная временем реализация с хорошим балансом между производительностью и качеством.

Как эталонная реализация градиентного бустинга, она служит хорошей точкой сравнения для более специализированных алгоритмов (CatBoost, LightGBM). Ее предсказуемость и стабильность важны для воспроизводимости результатов.

$
F_{m}(x) = F_{m-1}(x) + ν * γ_{m} * h_{m}(x)
$

Где γ_m — оптимальный вес, h_m(x) — новое дерево (базовый алгоритм), обучаемое на антиградиентах (псевдо-остатках), а ν (learning rate) — коэффициент скорости обучения.

4. **RandomForest (Случайный лес)**

RandomForest строит множество независимых решающих деревьев на случайных подвыборках данных и признаков, затем усредняет их предсказания. Это метод бэггинга (bootstrap aggregating), который снижает дисперсию по сравнению с отдельными деревьями.

RandomForest — отличный baseline-алгоритм, который редко сильно переобучается и хорошо работает "из коробки". Его способность оценивать важность признаков также полезна для анализа данных. В отличие от бустинга, он менее чувствителен к шуму и выбросам.

$
ŷ(x) = (1/B) * ∑_{b=1}^{B} T_b(x)
$

Где B — количество деревьев, T_b(x) — предсказание b-го дерева, обученного на бутстрап-выборке и подмножестве признаков.

5. **SVR (Support Vector Regression)**

SVR ищет функцию, которая отклоняется от истинных значений не более чем на заданную величину ε, максимизируя при этом запас. Это принципиально другой подход, основанный на теории статистического обучения и максимизации запаса, а не на минимизации ошибки.

SVR представляет класс алгоритмов, основанных на ядрах (kernel methods), что дополняет подходы на основе деревьев. Он особенно эффективен в случаях, когда истинная зависимость может быть описана небольшым числом опорных векторов. Также SVR имеет хорошие теоретические обоснования регуляризации.

Функция потерь (ε-insensitive loss) и условие оптимизации

$
L_ε(y, ŷ) = { 0, если |y - ŷ| ≤ ε; |y - ŷ| - ε, иначе }
$

Задача оптимизации (в упрощённом виде)

$
min (1/2 ||w||² + C * ∑_{i=1}^{n} (ξ_i + ξ_i*))
при условиях: |y_i - (w·φ(x_i) + b)| ≤ ε + ξ_i, ξ_i, ξ_i* ≥ 0
$

### Выбор метрик ML:
1. **MAE (Средняя абсолютная ошибка)** - для интерпретируемой оценки абсолютных ошибок
2. **RMSE (Среднеквадратичная ошибка)** - для чувствительной оценки больших ошибок
3. **R² (Коэффициент детерминации)** - для оценки доли объясненной дисперсии
4. **MAPE (Средняя абсолютная процентная ошибка)** - для относительной оценки ошибок

### Описание выбранных метрик:

- **MAE** - Средняя абсолютная ошибка

$
MAE = (1/n) * Σ|y_i - ŷ_i|
$

Пояснение:
1. Сумма абсолютных значений ошибок, деленная на количество наблюдений
2. Измеряется в тех же единицах, что и целевая переменная
3. Менее чувствительна к выбросам, чем RMSE

Средняя абсолютная ошибка MAE вычисляет среднее абсолютное значение разниц между предсказанными и фактическими значениями. Она измеряется в тех же единицах, что и целевая переменная, что делает её интуитивно понятной для интерпретации. Главное преимущество MAE в её устойчивости к выбросам, поскольку она не возводит ошибки в квадрат, однако это же свойство делает её менее чувствительной к крупным ошибкам, которые могут быть критичными в некоторых приложениях. Эта метрика хорошо подходит для случаев, когда все ошибки одинаково важны.

- **RMSE** - Среднеквадратичная ошибка

$
RMSE = √( (1/n) * Σ(y_i - ŷ_i)² ) = √(MSE)
$

Пояснение:
1. Квадратный корень из средней квадратичной ошибки
2. Измеряется в тех же единицах, что и целевая переменная
3. Более чувствительна к большим ошибкам (штрафует квадратично)

Среднеквадратичная ошибка RMSE представляет собой корень квадратный из средней квадратичной ошибки и, как и MAE, измеряется в единицах целевой переменной. Ключевое отличие - квадратичная природа этой метрики делает её более чувствительной к большим ошибкам, поскольку каждая ошибка возводится в квадрат перед усреднением. Это свойство особенно ценно в задачах, где крупные отклонения недопустимы или особенно дорого обходятся. RMSE часто используется как целевая функция при оптимизации моделей, но её интерпретация может быть менее интуитивной, чем у MAE, из-за влияния квадратичного преобразования.

- **R²** - Коэффициент детерминации

$
R² = 1 - (Σ(y_i - ŷ_i)²) / (Σ(y_i - ȳ)²)
$

где:

Σ(y_i - ŷ_i)² - сумма квадратов остатков

Σ(y_i - ȳ)² - общая сумма квадратов

y_i - истинное значение

ŷ_i - предсказанное значение

ȳ - среднее значение целевой переменной

i = 1, ..., n - количество наблюдений

Коэффициент детерминации R² показывает, какую долю дисперсии целевой переменной объясняет модель. Значения изменяются от -∞ до 1, где 1 означает идеальное соответствие модели данным, 0 - что модель предсказывает не лучше, чем простое среднее значение, а отрицательные значения указывают на то, что модель работает хуже этого базового уровня. Эта метрика особенно полезна для сравнения разных моделей на одной и той же задаче, поскольку она нормирована и позволяет оценить общую объясняющую способность модели независимо от масштаба данных. Однако R² не показывает абсолютную величину ошибок и может быть обманчиво высоким при наличии выбросов.

- **MAPE** - Средняя абсолютная процентная ошибка

$
MAPE = (1/n) * Σ|(y_i - ŷ_i) / y_i| * 100%
$

Пояснение:
1. Выражается в процентах
2. Позволяет сравнивать модели на данных разного масштаба
3. Не рекомендуется при значениях целевой переменной близких к нулю

Средняя абсолютная процентная ошибка MAPE выражает среднюю абсолютную ошибку в процентном отношении к фактическим значениям. Это делает её безразмерной метрикой, которая позволяет сравнивать качество моделей на задачах с разными масштабами данных - например, сравнивать точность прогноза цен на недвижимость и цен на акции. MAPE особенно популярна в бизнес-аналитике и экономическом прогнозировании, где ошибки традиционно выражают в процентах. Однако у этой метрики есть существенный недостаток: она не определена при нулевых значениях целевой переменной и может давать искажённые результаты при значениях, близких к нулю, а также асимметрично наказывает за заниженные и завышенные прогнозы.

Определим сетку гиперпараметров

```python
param_spaces = {

    "CatBoost": {
        "depth": ("int", 3, 10),
        "learning_rate": ("log", 1e-4, 0.3),
        "iterations": ("int", 200, 1500)
    },

    "LightGBM": {
        "num_leaves": ("int", 20, 150),
        "learning_rate": ("log", 1e-4, 0.3),
        "n_estimators": ("int", 200, 1500)
    },

    "GradientBoosting": {
        "n_estimators": ("int", 50, 500),
        "learning_rate": ("log", 1e-4, 0.3),
        "max_depth": ("int", 2, 6)
    },

    "RandomForest": {
        "n_estimators": ("int", 100, 700),
        "max_depth": ("int_none", 3, 25),
        "min_samples_split": ("int", 2, 10)
    },

    "SVR": {
        "C": ("log", 1e-2, 100),
        "gamma": ("log", 1e-4, 1)
    }
}
```

```python
def sample_param(trial, name, spec):
    t = spec[0]

    if t == "log":
        return trial.suggest_float(name, spec[1], spec[2], log=True)

    if t == "uniform":
        return trial.suggest_float(name, spec[1], spec[2])

    if t == "int":
        return trial.suggest_int(name, spec[1], spec[2])

    if t == "int_none":
        f = trial.suggest_categorical(name + "_flag", [True, False])
        return None if f else trial.suggest_int(name, spec[1], spec[2])

    raise ValueError("Unknown type:", t)
```

Импорт метрик из пула

```python
def optimize(model_name, X, y, trials=5):
    Model = models[model_name]
    space = param_spaces[model_name]

    best_k = 8

    def objective(trial):
        params = {k: sample_param(trial, k, v) for k, v in space.items()}

        if model_name in ("CatBoost"):
            model = Model(**params, verbose=False)
        elif model_name in ("LightGBM"):
            model = Model(**params, verbose=-1)
        else:
            model = Model(**params)

        pipe = Pipeline([
            ("scaler", StandardScaler()),
            ("model", model)
        ])

        cv = KFold(n_splits=best_k, shuffle=True, random_state=42)

        scores = cross_val_score(
            pipe, X, y,
            cv=cv,
            scoring="neg_root_mean_squared_error",
            n_jobs=-1
        )

        return -scores.mean()

    study = optuna.create_study(direction="minimize")
    study.optimize(objective, n_trials=trials)

    return study.best_params, study.best_value
```

# Обучение моделей ML и подбор гиперпараметров <a id='toc4_'></a>

```python
def clean_params(p):
    return {k: v for k, v in p.items() if not k.endswith("_flag")}

trained = {}
best_params = {}
best_scores = {}

for name in models:
    print("\nOptimizing", name)
    params, score = optimize(name, X_train, y_train, trials=5)

    params = clean_params(params)
    best_params[name] = params
    best_scores[name] = score

    Model = models[name]

    if name in ("CatBoost"):
        model = Model(**params, verbose=False)
    elif name in ("LightGBM"):
        model = Model(**params, verbose=-1)
    else:
        model = Model(**params)

    pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("model", model)
    ])

    pipe.fit(X_train, y_train)
    trained[name] = pipe
```

Например, CatBoost склонен несколько выше оценивать категориальные или сложные признаки, связанные с ядром и эмульсией, в то время как LightGBM, благодаря своей эффективности, может более резко отделить топ-признаки от всех остальных. GradientBoosting показывает схожую с CatBoost структуру, а RandomForest, как наиболее демократичный метод, даёт более сглаженное распределение важности, но при этом всё равно чётко подтверждает лидерство топ-признаков. Совокупность этих результатов даёт не только чёткий приоритетный список для инженера признаков или упрощения модели, но и глубокую перекрёстную валидацию, укрепляющую доверие к выводам о том, что именно параметры, связанные с давлением истоков и потоком энергий, являются основными драйверами целевого показателя в данной магико-технической системе.

```python
import shap
shap.initjs()

best_model_name = metrics_df.iloc[0]["model"]
best_model = trained[best_model_name].named_steps["model"]

explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X_train)

shap.summary_plot(shap_values, X_train)
```

Здесь ключевым выводом является то, что важнейший для всех предыдущих моделей признак «Давление выдоха Истока» оказывает неоднозначное влияние: его высокие значения как увеличивают, так и уменьшают целевую переменную (значения SHAP разбросаны по обе стороны от нуля). Это указывает на сложную, нелинейную или контекстную роль этого параметра, которую не раскрывала простая важность.

Истинным драйвером, стабильно увеличивающим выход модели (положительные SHAP значения), является признак «Приток Силы Потока». Именно его высокие значения имеют наиболее выраженный и последовательный положительный вклад. Следующие за ним «Эффективность Ядра» и «Полная Сила Потоков» также демонстрируют чёткий положительный тренд. Таким образом, SHAP-анализ смещает фокус: если агрегированные метрики важности указывали на «Давление» как на главный фактор, то анализ влияния показывает, что ключевым положительным рычагом является именно «Приток Силы Потока». Это критически важное уточнение для интерпретации модели: наиболее важный признак не обязательно является тем, что целенаправленно увеличивает результат, в то время как менее значимый в глобальном смысле признак может быть ключевым драйвером роста.
